```python
from sklearn.preprocessing import LabelEncoder
weathertype=LabelEncoder()
df['Weather Type']=weathertype.fit_transform(df['Weather Type'])
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot=OneHotEncoder(sparse_output=False,drop='first')
res=onehot.fit_transform(df[['Cloud Cover','Season','Location']])
df=df.drop(columns=['Cloud Cover','Season','Location'])
```

```python
res=pd.DataFrame(res)
res.columns=res.columns.astype(str)
```

```python
df=df.join(res)
```

```python
X=df.drop(columns='Weather Type')
y=df['Weather Type']
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
x_scaled=minmax.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
```

```python
models=[KNeighborsClassifier(),SVC(),GaussianNB()]
for model in models:
  print(model)
  model.fit(x_train,y_train)
  y_pred=model.predict(x_test)
  print(classification_report(y_test,y_pred))
  print('------------------------------------------')
```

```python
def weatherprediction():
  Temperature=int(input('enter tempreture'))
  Humidity=int(input('enter Humidity'))
  wind_speed=int(input('enter wind_speed'))
  Percipitation=int(input('enter Percipitation'))
  Atmosphere_pressure=int(input('enter Atmospheric Pressure'))
  UVindex=int(input('enter UV index'))
  Visibility=int(input('enter Visibility'))
  cloud=input('enter Cloud cover(overcast,partly,cloudy,clear,cloudy)')
  season=input('enter season(Winter,Spring,Autumn,Summer)')
  location=input('enter location(inland,mountain,coastal)')

  num=np.array([[Temperature,Humidity,wind_speed,Percipitation,Atmosphere_pressure,UVindex,Visibility]])
  var=np.array([[cloud,season,location]])

  x=onehot.transform(var)
  xx=np.hstack([num,x])
  xx_scaled=minmax.transform(xx)
  s=model.predict(xx_scaled)
  return weathertype.inverse_transform(s)


print(weatherprediction())
```

```python
import ipywidgets as widgets
from IPython.display import display
import numpy as np

# Function to perform the weather prediction
def weatherprediction_interactive():
    # Create input widgets
    temperature = widgets.FloatText(description="Temperature:")
    humidity = widgets.FloatText(description="Humidity:")
    wind_speed = widgets.FloatText(description="Wind Speed:")
    precipitation = widgets.FloatText(description="Precipitation:")
    atmospheric_pressure = widgets.FloatText(description="Atm Pressure:")
    uv_index = widgets.FloatText(description="UV Index:")
    visibility = widgets.FloatText(description="Visibility:")

    # Dropdown menus for categorical features
    cloud_cover = widgets.Dropdown(
        options=["overcast", "partly cloudy", "clear", "cloudy"],
        description="Cloud Cover:"
    )
    season = widgets.Dropdown(
        options=["Winter", "Spring", "Autumn", "Summer"],
        description="Season:"
    )
    location = widgets.Dropdown(
        options=["inland", "mountain", "coastal"],
        description="Location:"
    )

    # Button to trigger prediction
    button = widgets.Button(description="Predict Weather")
    output = widgets.Output()

    # Function to handle button click and make prediction
    def on_button_click(b):
        with output:
            output.clear_output()  # Clear previous output
            # Collect input values
            num = np.array([[temperature.value, humidity.value, wind_speed.value,
                             precipitation.value, atmospheric_pressure.value, uv_index.value, visibility.value]])
            var = np.array([[cloud_cover.value, season.value, location.value]])

            # Encode and scale the input
            x_encoded = onehot.transform(var)
            xx = np.hstack([num, x_encoded])
            xx_scaled = minmax.transform(xx)

            # Make prediction
            weather_type_encoded = model.predict(xx_scaled)
            weather_type = weathertype.inverse_transform(weather_type_encoded)

            # Display result
            print(f"Predicted Weather Type: {weather_type[0]}")

    button.on_click(on_button_click)

    # Display all widgets
    display(temperature, humidity, wind_speed, precipitation, atmospheric_pressure, uv_index,
            visibility, cloud_cover, season, location, button, output)

# Call the interactive function
weatherprediction_interactive()
```
